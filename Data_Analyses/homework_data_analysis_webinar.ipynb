```python
#Exercise 4: Try to plot the below chart using only df_apps dataframe and in 2 lines of code.

# data analysis and wrangling
import pandas as pd
import numpy as np

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df_reviews = pd.read_csv("../input/google-play-store-apps/googleplaystore_user_reviews.csv")
df_apps = pd.read_csv("../input/google-play-store-apps/googleplaystore.csv")

df_rating_per_category_ordered = df_apps[['Category','Rating']].groupby('Category').mean().iloc[1:].sort_values("Rating")
df_rating_per_category_ordered.plot(kind='bar', figsize=(10,5)).set_ylim(3.5,5)
```
